# Explanatory Data Visualization



```python
import pandas as pd
import matplotlib.pyplot as plt

top20_deathtoll = pd.read_csv('top20_deathtoll.csv')

# OOI uses methods to create and modify plots
# To create a graph using the OO interface, we use the plt.subplots() function, 
# which generates an empty plot and returns a tuple of two objects:
# fig (canvas) and an array of the actual plots
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'], 
       height=0.45) # To make the bars less thick, we can use the height parameter inside the Axes.barh() method.

for position in ['left', 'right', 'bottom', 'top']:
    ax.spines[position].set_visible(False) # removes the graph axes (also called spines)
    
# To remove some of the x-tick labels, we use the Axes.set_xticks method. 
# Below, we only keep the labels 0, 100000, 200000, and 300000.
ax.set_xticks([0, 150000, 300000])

ax.set_yticklabels([]) # to left-align the y-tick labels we will first remove them

country_names = top20_deathtoll['Country_Other']
for i, country in zip(range(20), country_names):
    ax.text(x=-80000, y=i-0.15, s=country)

ax.set_xticklabels(['0', '150,000', '300,000']) # add a comma to y-tick labels
ax.xaxis.tick_top() # move the tick labels at the top of the graph 
ax.tick_params(top=False, left=False) # remove the ticks
ax.tick_params(axis='x', colors='grey') # color the x-tick labels in grey so they don't stand out visually so much and color the bars in a shade of red.

# add a title and a subtitle, we're going to use the Axes.text() method. x and y are the coordinates that give the position of the text.
ax.text(x=-80000, y=23.5, s='The Death Toll Worldwide Is 1.5M+', size=17, weight='bold')
ax.text(x=-80000, y=22.5, s='Top 20 countries by death toll (December 2020)', size=12)

# Readers who explore the graph will try to determine the approximate death toll for each country. To help them, we're going to draw a vertical line below 
# the 150,000 value. To do that, we use the Axes.axvline(x) method, where x is the x-coordinate where the line begins:
ax.axvline(x=150000, ymin=0.045, c='grey', alpha=0.5)
plt.show()
```
